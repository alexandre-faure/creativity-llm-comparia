## 4.1. Paradoxe du juge

### La diversité comme force : l'évaluation distribuée a ses mérites
La littérature sur la CAT (*Consensual Assessment Technique*) montre que l'accord entre juges suffit comme base d'évaluation; ils n'ont pas forcément à savoir pourquoi une production est créative, du moment qu'ils s'accordent sur le fait qu'elle l'est ([SciSpace](https://scispace.com/pdf/amabile-s-consensual-assessment-technique-why-has-it-not-4o93phkmls.pdf)). Dans ce cadre, la masse et la diversité des utilisateurs de compar:IA sont une vraie force : elles permettent d'agréger des jugements sur des domaines très variés (humour, code, écriture, raisonnement), là où un panel d'experts serait nécessairement spécialisé et donc partiel.

Le problème de qui sont les "bons" juges est d'ailleurs ouvert dans la littérature : pour des domaines comme les légendes de dessins animés, il n'est pas clair qu'il existe un groupe d'experts reconnu ([Springer](https://link.springer.com/chapter/10.1007/978-3-319-95498-1_3)). La créativité dans la conversation quotidienne, qui est précisément ce que compar:IA mesure, appartient clairement à cette catégorie.

### La diversité comme limite : mais de quel type ?

La question originale sous-entend que des utilisateurs "sans les bonnes structures cognitives" rateraient des réponses véritablement créatives. C'est une affirmation à prendre avec des pincettes. Ce qui est plus documenté, c'est que l'évaluation de la créativité est médiée par des biais psychosociaux: notamment la proximité sociale et culturelle entre le juge et ce qui est évalué ([Theses.fr](https://theses.fr/2022MON30085)). Autrement dit, le problème n'est pas que les utilisateurs "manquent de structures cognitives" : c'est qu'ils ont chacun des cadres de référence situés qui favorisent certains types de créativité (souvent ceux qui ressemblent à leur propre univers culturel).

Ce biais existe tout autant chez les experts, voire plus, puisqu'un panel d'experts partage souvent les mêmes normes disciplinaires, ce qui peut conduire à sur-valoriser l'originalité intra-domaine et ignorer des formes de créativité transversales ou populaires.

### Le panel d'experts : gold standard ou standard biaisé ?
La CAT est modélisée sur la façon dont la créativité est jugée au plus haut niveau dans la plupart des domaines par des panels d'experts. Quand le prix Nobel de littérature est décerné, ce n'est pas au vote populaire ([ScienceDirect](https://www.sciencedirect.com/topics/psychology/consensual-assessment-technique)). C'est un argument fort. Mais il naturalise une convention sociale : le jugement des experts n'est pas neutre, il reproduit les hiérarchies culturelles existantes.

Des études sur des compositions musicales d'enfants montrent d'ailleurs que la fiabilité inter-juges varie énormément selon le type d'experts : les compositeurs avaient une fiabilité de seulement $.04$, contre $.73$ pour les théoriciens ([SAGE](https://journals.sagepub.com/doi/10.2307/3345709)), ce qui montre que "expert" ne veut pas forcément dire "accord".

### Méthodologie idéale : une approche à deux niveaux

Plutôt qu'opposer crowd et experts, la méthodologie idéale combinerait :

1. L'évaluation crowd de compar:IA pour sa couverture large, sa diversité culturelle et sa validité écologique (les gens évaluent des réponses à leurs propres questions).
2. Un panel expert post-hoc stratifié sur des sous-ensembles ciblés (ex : réponses créatives en code, en poésie, en raisonnement), pour calibrer et auditer les jugements populaires, sans les remplacer.

Des travaux montrent d'ailleurs qu'on peut calibrer des juges non-experts sur des juges experts de façon très satisfaisante, avec des corrélations proches de $.91$ ([PubMed](https://pubmed.ncbi.nlm.nih.gov/16060417/)), ce qui suggère que les deux niveaux ne sont pas irréconciliables.
Cette approche a l'avantage de ne pas présupposer qu'il existe une bonne définition de la créativité, mais de traiter le jugement comme ce qu'il est : un acte social, situé, et toujours partiel.
